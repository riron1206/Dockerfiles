# 1. Data Repository

```python
from pycaret.datasets import get_data

index = get_data("index")
```

```python
data = get_data("juice")
```

# 2. Initialize Setup

```python
from pycaret.classification import *

clf1 = setup(
    data,
    target="Purchase",
    session_id=123,
    log_experiment=True,
    experiment_name="juice1",
)
```

# 3. Compare Baseline

```python
best_model = compare_models()
```

# 4. Create Model

```python
lr = create_model("lr")
```

```python
dt = create_model("dt")
```

```python
rf = create_model("rf", fold=5)
```

```python
models()
```

```python
models(type="ensemble").index.tolist()
```

```python
ensembled_models = compare_models(
    whitelist=models(type="ensemble").index.tolist(), fold=3
)
```

# 5. Tune Hyperparameters

```python
tuned_lr = tune_model(lr)
```

```python
tuned_rf = tune_model(rf)
```

# 6. Ensemble Model

```python
bagged_dt = ensemble_model(dt)
```

```python
boosted_dt = ensemble_model(dt, method="Boosting")
```

# 7. Blend Models

```python
blender = blend_models(estimator_list=[boosted_dt, bagged_dt, tuned_rf], method="soft")
```

# 8. Stack Models

```python
stacker = stack_models(estimator_list=[boosted_dt, bagged_dt, tuned_rf], meta_model=rf)
```
