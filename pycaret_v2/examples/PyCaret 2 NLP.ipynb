# 1. Loading Dataset

```python
from pycaret.datasets import get_data

data = get_data("kiva")
```

# 2. Initialize Setup

```python
from pycaret.nlp import *

nlp1 = setup(
    data,
    target="en",
    session_id=123,
    log_experiment=True,
    log_plots=True,
    experiment_name="kiva1",
)
```

# 3. Create Model

```python
models()
```

```python
lda = create_model("lda")
```

```python
nmf = create_model("nmf", num_topics=6)
```

# 4. Assign Labels

```python
lda_results = assign_model(lda)
lda_results.head()
```

# 5. Analyze Model

```python
plot_model(lda)
```

```python
plot_model(lda, plot="bigram")
```

```python
plot_model(lda, plot="tsne")
```
