# MMDetectionで物体認識試す
- MMDetection: pytorchで物体認識を行うライブラリ
- 参考: https://www.soliton-cyber.com/blog/mmdet

```python
import mmdet
from mmdet.apis.inference import init_detector, inference_detector, show_result_pyplot
```

```python
# 学習済みモデルを取得。リンクを辿り、学習済みモデルのURLを調べて、wgetで取得
!mkdir -p ./checkpoints
# https://github.com/open-mmlab/mmdetection/tree/master/configs/faster_rcnn
!wget -P ./checkpoints https://download.openmmlab.com/mmdetection/v2.0/faster_rcnn/faster_rcnn_r50_fpn_1x_coco/faster_rcnn_r50_fpn_1x_coco_20200130-047c8118.pth
```

```python
# ソースの中のconfigファイルと先に取得した重みファイルを指定してモデルを作成します。deviceには'cuda:0'を指定
config = f'{MMDET_DIR}/configs/faster_rcnn/faster_rcnn_r50_fpn_1x_coco.py'
checkpoint = f'./checkpoints/faster_rcnn_r50_fpn_1x_coco_20200130-047c8118.pth'
device = 'cuda:0'
model = init_detector(config, checkpoint, device=device)
```

```python
# ソースの中の画像を利用して物体認識を実行し、結果を表示
import warnings
warnings.simplefilter('ignore')
result = inference_detector(model, f'{MMDET_DIR}/demo/demo.jpg')
show_result_pyplot(model, f'{MMDET_DIR}/demo/demo.jpg', result, score_thr=0.7)
```

```python
# Mask R-CNNの学習済みの重みファイルを取得
!mkdir -p ./checkpoints
# https://github.com/open-mmlab/mmdetection/tree/master/configs/mask_rcnn
!wget -P ./checkpoints https://download.openmmlab.com/mmdetection/v2.0/mask_rcnn/mask_rcnn_r50_fpn_1x_coco/mask_rcnn_r50_fpn_1x_coco_20200205-d4b0c5d6.pth
```

```python
# ソースの中のconfigファイルと先に取得した重みファイルを指定してモデルを作成して推論
config = f'{MMDET_DIR}/configs/mask_rcnn/mask_rcnn_r50_fpn_1x_coco.py'
checkpoint = f'./checkpoints/mask_rcnn_r50_fpn_1x_coco_20200205-d4b0c5d6.pth'
device = 'cuda:0'
model = init_detector(config, checkpoint, device=device)

result = inference_detector(model, f'{MMDET_DIR}/demo/demo.jpg')
show_result_pyplot(model, f'{MMDET_DIR}/demo/demo.jpg', result, score_thr=0.7)
```
