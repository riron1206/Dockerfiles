```python
import pandas as pd
import numpy as np

from pycaret.datasets import get_data
dataset = get_data("diamond")

from pycaret.regression import *
```

```python
help(setup)
```

```python
# setup()に追加optionなし

exp = setup(dataset,
            session_id=123,
            target="Price")
display(exp[0].head())
```

```python
# 「silent=True」を渡すことで、型推定の確認をスキップ

exp = setup(dataset,
            session_id=123,
            silent = True,
            target="Price"
           )

#exp は (x, y, train_x, val_x, train_y, val_y, session_id, sklearn_pipeline)のタプル
display(exp[0].head())
```

```python
# 型指定とかいろいろ

exp = setup(dataset,
            session_id=123,
            silent = True,
            target="Price",
            train_size=0.9,  # train:validatio=0.9:0.1
            categorical_features=["Cut", "Color", "Clarity", "Polish", "Symmetry", "Report"], 
            numeric_features=["Carat Weight"],
            normalize=True,  # 各numeric_featuresを正規化
            normalize_method='zscore',  # 正規化の方法は標準化
            numeric_imputation='mean',  # 各numeric_featuresの欠損埋めの方法を指定。平均値で埋める
            categorical_imputation='constant',  # カテゴリデータに対する欠損埋めの方法を指定。「not_available」という文字列で埋める
            bin_numeric_features=['Carat Weight'],  # ビン化したい数値データ特徴量のカラム名をリストで指定
            remove_outliers=True,  # PCAを使用して外れ値除去
            outliers_threshold=0.05,  # 外れ値の割合を指定。デフォルトの0.05を指定すると、分布の裾の両側の値の0.025%が除去
            remove_multicollinearity=True,  # マルチコ（多重共線性）が発生しうる特徴量を除去
            multicollinearity_threshold=0.9,  # 本パラメータで定義されたしきい値よりも高い相互相関を持つ変数が削除
            create_clusters=True,  # 各特徴量を用いたクラスタリングを行い、各レコードのクラスラベルを新たな特徴量として追加
            cluster_iter=20,  # クラスタ数を決める際の反復数を指定。default=20
            ignore_low_variance=True,  # 統計的に有意ではない分散を持つ特徴量を削除
            polynomial_features=True,  # 全ての数値データ特徴量の多項式の組み合わせで新たな特徴量を生成
            polynomial_degree=2,  # 多項式特徴の次数を指定。default = 2
            transformation=True,  # 特徴量をスケーリング
            transformation_method='yeo-johnson',  # sklearn.preprocessing.PowerTransformer。Box-Cox変換っぽい正規分布に近づける変換
            transform_target=True,  # 目的変数をスケーリング
            transform_target_method='box-cox',  # Box-Cox変換
            pca=True,  # 特徴量の次元削減
            pca_method='linear',  # 'linear' : 主成分分析（線形）による次元削減
            pca_components=0.99,  # 次元削減後に残す特徴量の数/割合を指定。default = 0.99
            profile=True,  # pandas-profiling の結果表示
           )
display(exp[0].head())
```
